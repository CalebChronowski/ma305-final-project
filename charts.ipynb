```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from equations import equations

eq_i, eq_ii = equations[0], equations[1]
```

# equation i
$\int_{0}^{1} 4 \sqrt{1-x^2} dx$

```python
fig = plt.figure()
ax = fig.add_subplot

x = np.linspace(0, 1, 100) 
y = eq_i(x)

# Plot the surface
plt.plot(x, y)

plt.show()
```
