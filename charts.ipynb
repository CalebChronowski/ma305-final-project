# Part 1
## accuracy vs. iterations

```python
def get_data(integration_rule, fn, a, b, tol=0.00001):
    x, y = [], []
    n = 1
    error = 1
    pi = np.pi
    while error > tol:
        pi_estimate = integration_rule(fn, a, b, n)
        error = (pi - pi_estimate)/pi
        x.append(n)
        y.append(error)
        n+=1
    return np.array(x), np.array(y)
```

```python
simpsons = get_data(simpsons_rule, eq_i, 0, 1, 0.001)

fig = plt.figure()
ax = fig.add_subplot

x, y = simpsons

# Plot the surface
plt.scatter(x, y)
plt.title(f"Simspons 1/3 rule with equation i\nerror vs iterations\nN={x.shape[0]}")
plt.show()
```

```python
trapezoid = get_data(trapezoid_rule, eq_i, 0, 1, 0.001)

fig = plt.figure()
ax = fig.add_subplot

x, y = trapezoid

# Plot the surface
plt.scatter(x, y)
plt.title(f"Trapezoid rule with equation i\nerror vs iterations\nN={x.shape[0]}")
plt.show()
```
